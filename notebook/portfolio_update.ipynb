```python
# Cell 2 — Authenticate + load secrets
#
# One-time setup: add these five keys via the lock icon (Secrets) in the left sidebar:
#   DRIVE_UPLOAD_FOLDER_ID, DRIVE_PROCESSED_FOLDER_ID, DRIVE_FAILED_FOLDER_ID,
#   DRIVE_CACHE_FOLDER_ID, SPREADSHEET_ID

import os
from google.colab import auth, userdata

auth.authenticate_user()
import google.auth
credentials, _ = google.auth.default()

os.environ["DRIVE_UPLOAD_FOLDER_ID"]    = userdata.get("DRIVE_UPLOAD_FOLDER_ID")
os.environ["DRIVE_PROCESSED_FOLDER_ID"] = userdata.get("DRIVE_PROCESSED_FOLDER_ID")
os.environ["DRIVE_FAILED_FOLDER_ID"]    = userdata.get("DRIVE_FAILED_FOLDER_ID")
os.environ["DRIVE_CACHE_FOLDER_ID"]     = userdata.get("DRIVE_CACHE_FOLDER_ID")
os.environ["SPREADSHEET_ID"]            = userdata.get("SPREADSHEET_ID")

print("Auth OK — env vars loaded from Colab secrets")
```

```python
# Cell 3 — Run the portfolio update
from portfolio.runner import run_update
run_update(credentials=credentials)
```
